1- Load the Gapminder dataset

```python
import pandas as pd
```

```python
data = pd.read_csv("./data/gapminder.tsv", delimiter='\t')
```

```python
data.head()
```

2- Relate per capita GDP to life expectancy using a scatter plot

```python
sns.relplot(x='gdpPercap', y='lifeExp', data=data)
```

2- Use color to differentiate between the different years

```python
sns.relplot(x='gdpPercap', y='lifeExp', hue='year', data=data)
```

4- Can you account for population size? Try changing the size of the dots

```python
sns.relplot(x='gdpPercap', y='lifeExp', hue='year', size='pop', sizes=(15,500), data=data)
```

5- Use facets to break this plot into one plot per year

```python
sns.relplot(x='gdpPercap', y='lifeExp', col='year', data=data)
```

6- how does life expectancy vary per continent?

```python
sns.catplot(x='continent', y='lifeExp', kind='boxen', data=data)
```

7- how did life expectancy change over time?

```python
sns.catplot( x='lifeExp', y='continent', col='year', kind='boxen', data=data)
```
