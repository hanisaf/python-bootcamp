```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
# print graphs in the document
%matplotlib inline
```

1- Load the Gapminder dataset

```python
data = pd.read_csv('./data/gapminder.tsv', delimiter='\t')
```

```python
data.head()
```

2- Relate per capita GDP to life expectancy using a scatter plot

```python
sns.relplot(x="gdpPercap", y="lifeExp", data=data)
```

2- Use color to differentiate between the different years

```python
sns.relplot(x="gdpPercap", y="lifeExp", hue="year", data=data)
```

```python
sorted(data.year)[-10:-1]
```

4- Can you account for population size? Try changing the size of the dots

```python
sns.relplot(x="gdpPercap", y="lifeExp", hue="year", size="pop", data=data)
```

```python
sns.relplot(x="gdpPercap", y="lifeExp", hue="year", size="pop", sizes=(15,200), data=data)
```

5- Use facets to break this plot into one plot per year

```python
sns.relplot(x="gdpPercap", y="lifeExp", col="year", size="pop", sizes=(15,200), data=data)
```

5- how does life expectancy vary per continent?

```python
sns.catplot(x="lifeExp", y="continent", kind="boxen",
            data=data);
```

```python
sns.catplot(x="lifeExp", y="continent", kind="boxen", col="year",
            data=data);
```

6- how did life expectancy change over time?

```python
sns.relplot(x="lifeExp", y="year", data=data)
```

```python
sns.jointplot(x="lifeExp", y="year", data=data, kind="kde");
```

```python
sns.distplot(data["lifeExp"])
```

```python
for year in set(data["year"]):
    subset = data[data.year==year]
    sns.distplot(subset["lifeExp"])
```
